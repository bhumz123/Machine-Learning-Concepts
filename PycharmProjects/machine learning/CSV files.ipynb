```python
# SEP Parameter
# by default its comma
# \t represents tab values ie space
pd.read_csv("movie_titles_metadata.tsv", sep="\t")    
```

```python
# to give the column names
# we use the names
# names is used to give the column names 

pd.read_csv("movie_titles_metadata.tsv", sep="\t",names=["sno","movie name",
"yearofrelease","imb rating","votes","genre" ])    
```

```python
# index_col parameter
# this is used to convert a column(not required) into a row 

pd.read_csv("movie_titles_metadata.tsv", sep="\t",names=["sno","movie name",
"yearofrelease","imb rating","votes","genre" ], index_col='sno')  
```

```python
# # header parameter
# # when names of attributes are in the row , to get them as attributes we use this
 pd.read_csv("", header=1)
```

```python
# usecols parameter
# used when we only want specific attributes

pd.read_csv("", usecols=[] 
```

```python
#squeeze parameters
# used when you want only one attribute , it becomes a series rather than a datafram
pd.read_csv("aug_train.csv",usecols=['enrollee_id'], squeeze=True)
```

```python
# skips rows
pd.read_csv("aug_train.csv",skiprows=[0,9])
```

```python
# nrows - when we want only specific number of rows
pd.read_csv("aug_train.csv",nrows=100)
```

```python
# skip bad lines, when there are more than required attributes by mistake
## error_bad_lines
pd.read_csv("aug_train.csv",error_bad_lines=True)
```

```python
# dtype
# to change the dtype of an attribute
pd.read_csv("aug_train.csv",dtype={'target':int})
```

```python
# handling dates
# parse_date: used to convert dtype into date time to use the date time parameters
pd.read_csv("IPL Matches 2008-2020.csv", parse_dates=['date']).info()
```

```python
def rename(name):
    if name=="Kolkata Knight Riders":
        return ("KKR")
    else:
        return name
```

```python
#convertors
pd.read_csv("IPL Matches 2008-2020.csv",converters={'team1':rename})
```

```python
# chuns
# to lead large datasets in chunks
dfs=pd.read_csv("aug_train.csv",chunksize=5000)
for i in dfs:
    print(i.shape)
```
